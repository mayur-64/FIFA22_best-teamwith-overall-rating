As above, players with higher overall ratings are getting repeated in the above logic, to solve it. We have created a new 
funtion which will take input one list and a variable and check weather that variable is present in the list or not.

```python
def check(list1,val):
    for x in list1:
        if val==x:
            return False
    return True
```

# Best National team 

```python
def best_nation_team(nationality):
    group = fifa2.groupby("Nationality")
    nation_team=group.get_group(nationality)
    nation_team.sort_values(by=["Overall"], axis=0,ascending=False, inplace = True)
    temp=pd.DataFrame({"ID_NO":[],"Name":[],"Age":[],"Nationality":[],"Overall":[],"Value":[],"Best Position":[],"Club":[]})
    for f in range(len(f433)):
        #print(f433[f])
        for d in range(len(nation_team)):
            if(f433[f]==nation_team.iloc[d][6] and check(temp["ID_NO"],nation_team.iloc[d][0])):
                temp = temp.append({"ID_NO":nation_team.iloc[d][0],"Name":nation_team.iloc[d][1],"Age":nation_team.iloc[d][2],
                                    "Nationality":nation_team.iloc[d][3],'Overall':nation_team.iloc[d][4],
                                    "Value":nation_team.iloc[d][5],"Best Position":nation_team.iloc[d][6],
                                    "Club":nation_team.iloc[d][7]},ignore_index=True)
                break
    if len(temp)< 11:
        print("Invalid Formation",f433)
    else : print(f433)
    print(nationality,"\nAverage age of the team",temp["Age"].mean())
    print("Overall team Rating",temp["Overall"].mean())
    return temp
```

# Best Club team

```python
def best_club_team(club_name):
    group = fifa2.groupby("Club")
    club_team=group.get_group(club_name)
    club_team.sort_values(by=["Overall"], axis=0,ascending=False, inplace = True)
    temp=pd.DataFrame({"ID_NO":[],"Name":[],"Age":[],"Nationality":[],"Overall":[],"Value":[],"Best Position":[],"Club":[]})
    for f in range(len(f442)):
        #print(f433[f])
        for d in range(len(club_team)):
            if(f442[f]==club_team.iloc[d][6] and check(temp["ID_NO"],club_team.iloc[d][0])):
                temp = temp.append({"ID_NO":club_team.iloc[d][0],"Name":club_team.iloc[d][1],"Age":club_team.iloc[d][2],
                                    "Nationality":club_team.iloc[d][3],'Overall':club_team.iloc[d][4],
                                    "Value":club_team.iloc[d][5],"Best Position":club_team.iloc[d][6],
                                    "Club":club_team.iloc[d][7]},ignore_index=True)
                break
    
    if len(temp)<11:
        print("Invalid Formation\n",f442)
    else: print(f442)
    print(club_name,"\nAverage age of the team",temp["Age"].mean())
    print("Overall team Rating",temp["Overall"].mean())
    return temp
```
